# Iris Visualization

### Importing Liabraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
df = pd.read_csv(r'iris.csv')
df.head(3)
```

### Data Visualization

```python
sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=df,
    palette = "viridis"
)

plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.histplot(df["petal_length"], bins=30)

plt.title("Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

plt.show()
```

```python
sns.boxplot(x="species", y="petal_width", data=df)

plt.title("Petal Width by Species")

plt.show()
```

### Conclution 

In this task, the Iris dataset was explored and visualized using Pandas, Matplotlib, and Seaborn. The dataset contains information about different flower species including setosa, versicolor, and virginica.

The scatter plot showed the relationship between sepal length and sepal width. It helped identify how different species are grouped based on their measurements. Setosa flowers were clearly separated from the other species.

The histogram displayed the distribution of petal length values. Most flowers had petal lengths between 1 cm and 6 cm, showing variation among species.

The box plot compared petal width across species and highlighted differences in spread and possible outliers. Virginica had the largest petal width, while setosa had the smallest.
